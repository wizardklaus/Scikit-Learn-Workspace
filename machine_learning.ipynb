```python
# Machine Learning House Prices Prediction
# 데이터 불러오기
#pip install scikit-learn
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

df = pd.read_csv("https://raw.githubusercontent.com/rasbt/python-machine-learning-book/master/code/datasets/housing/housing.data", delimiter=r"\s+", 
                names = ["CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM", "AGE", "DIS", "RAD", "TAX", "PTRATIO", "B", "LSTAT", "MEDV"])
df.head() #데이터의 꼴을 본다
df.shape #데이터가 몇차원으로 이루어져있는지 확인한다.
df[pd.isnull(df).any(axis=1)] #isnull은 null값이 존재하는지 검사해준다. axis는 0은 행 1은 열이다.
X = df[["CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM", "AGE", "DIS", "RAD", "TAX", "PTRATIO", "B", "LSTAT"]]
y = df[["MEDV"]]
#x축과 y축에 값을 넣어준다.
```

```python
#Train Test Set 나누기
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=5)
#학습할 데이터와 테스트할 데이터를 나누는 부분, test_size는 테스트 데이터 비율, random_state는 seed값이다.
```

```python
#Gradient Descent 함수 생성
def gradientDescent(x,y,theta,alpha,n,numlterations):
    tmp=x.copy()
    tmp=(tmp-tmp.mean())/(tmp.max()-tmp.min()) #정규화를 하는 부분
    tmp.insert(0,"bias",1) #0번째 위치에 bias라는 이름의 변수(값:1)가 추가된다.
    #y와 x간의 관계를 구하면 절편값이 있어야하는데(bias) 그 값을 임의로 1로 정해준다.
    x=np.array(tmp)
    y=np.array(y).ravel() #ravel()은 구조를 없애준다.
    #bias 추가
    xTrans=x.transpose() #r에서의 t함수, 행과 열을 바꿔준다.
    for i in range(0,numlterations):
         hypothesis=np.dot(x,theta) #각 x값과 세타값들의 곱의 합을 구한다.
#          loss=#?
         #avg cost per example(the 2 in 2*n doesn't really matter here)
         #but to be consistent with the grandient, | include it)
         cost=sum((hypothesis-y)**2)/(2*n) #**는 제곱
         if(i%1000==0):
             print("Iteration %d | Cost: %f" % (i,cost))
         #avg gradient per example
         gradient=np.dot(xTrans,(hypothesis-y))/n
         #update
         theta=theta-alpha*gradient
    return theta
#tmp 부분 설명, 함수가 모델과 실제 y값과의 차이를 나타내는 함수이다. 잔차의 합이 최소화되어야 하는 것을 
#해서.. 미분을 하면 기울기가 나오면 그 기울기 값으로 뺀다(x=x-theta) 그래서 U자 그래프가 조금씩 움직여서
#0에 다다르게 한다.
```

```python
#예측 함수 생성
def predict(x,theta):
    return np.dot(x,theta)
```

```python
#학습
x=X_train
y=y_train
n,m=np.shape(x)
numIterations=300000
alpha=0.01
theta=np.ones(m+1)
theta=gradientDescent(x,y,theta,alpha,n,numIterations)
print(theta)
```

```python
#scikit-learn의 선형회귀식과 비교
#scikit-learn의 회귀식 모습
from sklearn import linear_model
reg=linear_model.LinearRegression()
X_train_norm=(X_train-X_train.mean())/(X_train.max()-X_train.min())
reg.fit(X_train_norm,y_train)

#결과 확인
print(reg.intercept_)
print(reg.coef_)
print(reg.predict([[0.03237,0.0,2.18,0,0.458,6.998,45.8,6.0622,3,222.0,18.7,394.63,2.94]]))
```

```python
#gradient descent 의 회귀식 결과 확인
print(theta)
print(predict([0.03237,0.0,2.18,0,0.458,6.998,45.8,6.0622,3,222.0,18.7,394.63,2.94],theta))
```
